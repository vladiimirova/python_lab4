```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

# Завантаження даних з Excel
energy_data = pd.read_excel('En_In.xls', skiprows=17)  # Пропускаємо верхні 17 рядків, щоб пропустити заголовки

# Виключаємо перші два стовпці та змінюємо назви
energy_data = energy_data.drop(columns=['Unnamed: 0', 'Unnamed: 1'])
energy_data.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

# 2. Перетворення "Energy Supply" в гДж (1 ПДж = 1,000,000 ГДж), заміна "....." на np.NaN
energy_data['Energy Supply'] = pd.to_numeric(energy_data['Energy Supply'], errors='coerce') * 1e-6

# 3. Перейменування країн
energy_data['Country'] = energy_data['Country'].replace({
    "Republic of Korea": "South Korea",
    "United States of America20": "United States",
    "United Kingdom of Great Britain and Northern Ireland19": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
})

 # 4. Видалення цифр і дужок з назв країн
energy_data['Country'] = energy_data['Country'].apply(lambda x: re.sub(r'(\d+|\(.*\))', '', str(x)).strip())

# Виведення випадкових 5 рядків для перевірки
print("Завдання 1-4 (випадкові рядки):")
print(energy_data.sample(5))  # Вивести 5 випадкових рядків для перевірки

# Завантаження даних про ВВП
gdp_data = pd.read_csv('gpd.csv', skiprows=4)

# Перейменовуємо країни
gdp_data['Country Name'] = gdp_data['Country Name'].replace({
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
})

# Перейменування стовпця 'Country Name' на 'Country'
gdp_data.rename(columns={'Country Name': 'Country'}, inplace=True)

# Виведення для завдання 5: частина даних після перейменування
print("\nЗавдання 5: Столбцы в gdp_data після перейменування:")
print(gdp_data.columns)  # Виводимо назви стовпців

# Виведення таблиці для першої країни
print("\nТаблиця для першої країни:")
print(gdp_data.head(1))  # Виводимо перший рядок таблиці

# 7. Приєднання трьох наборів даних до нового набору даних

# Фільтруємо 15 найкращих країн за рейтингом Scimagojr (Rank від 1 до 15)
scimagojr_top15 = scimagojr_data[scimagojr_data['Rank'] <= 15]

# Фільтруємо дані про ВВП для років 2006-2015
gdp_data_filtered = gdp_data[['Country', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]

print("Список країн у energy_data:")
print(energy_data['Country'].unique())

# Об'єднуємо дані: energy_data, gdp_data і scimagojr_data за країною
merged_data = energy_data.merge(gdp_data_filtered, on='Country', how='inner')
merged_data = merged_data.merge(scimagojr_top15, on='Country', how='inner')

# Встановлюємо країни як індекс
merged_data.set_index('Country', inplace=True)

# Перевіряємо, чи всі стовпці на місці і в правильному порядку
desired_columns = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',
                   'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', 
                   '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']

# Перепорядковуємо стовпці у потрібний порядок
merged_data = merged_data[desired_columns]

# Виведення результату
print("\nЗавдання 7: Після об'єднання та перестановки стовпців:")
print(merged_data.head())  # Перевіряємо перші кілька рядків


# Завдання 8: середній ВВП за останні 10 років
def average_gdp(data):
    gdp_columns = [str(year) for year in range(2006, 2016)]
    return data[gdp_columns].mean(axis=1).sort_values(ascending=False)

# Викликаємо функцію для завдання 8
avg_gdp = average_gdp(merged_data)
print("\nЗавдання 8: Середній ВВП за останні 10 років (відсортовано за спаданням):")
print(avg_gdp)

# Завдання 9: зміна ВВП для країни з 5-м середнім ВВП
def gdp_change(data):
    avg_gdp = average_gdp(data)
    fifth_country = avg_gdp.index[4]
    change = data.loc[fifth_country, '2015'] - data.loc[fifth_country, '2006']
    return (fifth_country, change)

# Викликаємо функцію для завдання 9
gdp_diff = gdp_change(merged_data)
print("\nЗавдання 9: Зміна ВВП за останні 10 років для країни з 5-м середнім ВВП:")
print(gdp_diff)

# Завдання 10: максимальний відсоток поновлюваної енергії
def max_renewable(data):
    max_renewable_country = data['% Renewable'].idxmax()
    max_renewable_value = data.loc[max_renewable_country, '% Renewable']
    return (max_renewable_country, max_renewable_value)

# Викликаємо функцію для завдання 10
max_renewable_country = max_renewable(merged_data)
print("\nЗавдання 10: Країна з найбільшим відсотком поновлюваної енергії:")
print(max_renewable_country)

# Завдання 11: оцінка чисельності населення
def population_estimate(data):
    data['Population Estimate'] = data['Energy Supply'] / data['Energy Supply per Capita']
    sixth_country = data['Population Estimate'].idxmin()
    return (sixth_country, data.loc[sixth_country, 'Population Estimate'])

# Викликаємо функцію для завдання 11
population = population_estimate(merged_data)
print("\nЗавдання 11: Шоста країна за оцінкою чисельності населення:")
print(population)

# Завдання 12: кореляція між цитованими документами та енергозабезпеченням
def correlation(data):
    data['Citable documents per capita'] = pd.to_numeric(data['Citable documents'], errors='coerce') / pd.to_numeric(data['Population Estimate'], errors='coerce')
    data['Energy Supply per Capita'] = pd.to_numeric(data['Energy Supply per Capita'], errors='coerce')

    # Видалення рядків з NaN значеннями
    data_clean = data.dropna(subset=['Citable documents per capita', 'Energy Supply per Capita'])

    # Обчислення кореляції
    correlation_value = data_clean['Citable documents per capita'].corr(data_clean['Energy Supply per Capita'])
    return correlation_value

# Викликаємо функцію для завдання 12
correlation_value = correlation(merged_data)
print("\nЗавдання 12: Кореляція між цитованими документами на душу населення та енергозабезпеченням:")
print(correlation_value)

# Завдання 13: відновлювана енергія
def renewable_energy(data):
    median_renewable = data['% Renewable'].median()
    data['Above median renewable'] = np.where(data['% Renewable'] >= median_renewable, 1, 0)
    return data['Above median renewable'].sort_values(ascending=True)

# Викликаємо функцію для завдання 13
renewable_status = renewable_energy(merged_data)
print("\nЗавдання 13: Додавання стовпця для відновлюваної енергії:")
print(renewable_status)

# Завдання 14: групування за континентами
ContinentDict = {
    'China':'Asia',
    'United States':'North America',
    'Japan':'Asia',
    'United Kingdom':'Europe','Russian Federation':'Europe',
    'Canada':'North America',
    'Germany':'Europe',
    'India':'Asia',
    'France':'Europe',
    'South Korea':'Asia',
    'Italy':'Europe',
    'Spain':'Europe',
    'Iran':'Asia',
    'Australia':'Australia',
    'Brazil':'South America'
}

def group_by_continent(data):
    data['Continent'] = data.index.map(ContinentDict)
    continent_grouped = data.groupby('Continent')['Population Estimate'].agg(['size', 'sum', 'mean', 'std'])
    return continent_grouped

# Викликаємо функцію для завдання 14
continent_stats = group_by_continent(merged_data)
print("\nЗавдання 14: Групування країн за континентами:")
print(continent_stats)
print("Список країн у даних після об'єднання:")
print(merged_data.index.unique())
# Завдання 15: бульбашкова діаграма
def bubble_chart(data):
    # Заміна NaN значень в стовпці 'Continent' на 'Unknown'
    data['Continent'] = data['Continent'].fillna('Unknown')

    # Визначення розміру бульбашок на основі ВВП за 2015 рік (масштабоване для зручності)
    bubble_size = data['2015'] * 1e-9  # Ви можете змінити множник для налаштування розміру

    # Визначення кольору для кожного континенту
    continent_colors = data['Continent'].map({
        'Asia': 'red', 
        'North America': 'blue', 
        'Europe': 'green', 
        'Australia': 'yellow', 
        'South America': 'orange', 
        'Unknown': 'gray'  # Для невідомих континентів
    })

    # Побудова бульбашкової діаграми
    plt.figure(figsize=(12, 8))
    scatter = plt.scatter(data['Rank'], data['% Renewable'], s=bubble_size, c=continent_colors, alpha=0.6)

    # Додавання підписів до кожної бульбашки (назви країн)
    for i, country in enumerate(data.index):
        plt.text(data['Rank'].iloc[i], data['% Renewable'].iloc[i], country, fontsize=9, ha='center', va='center')

    # Заголовок та мітки
    plt.title('% Renewable vs Rank (Bubble size represents GDP in 2015)', fontsize=16)
    plt.xlabel('Rank (1 = Best)', fontsize=12)
    plt.ylabel('% Renewable', fontsize=12)

    # Показуємо графік
    plt.tight_layout()
    plt.show()

# Викликаємо функцію для завдання 15
bubble_chart(merged_data)
```
